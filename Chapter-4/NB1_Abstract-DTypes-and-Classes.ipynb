Abstract data types are implemented using **classes**

```python
# Abstraction for the class of integer sets
class IntSet(object): #PASSING object TO IntSet() INDICATES IntSet() IS A SUBCLASS OF object
    """An intSet is a set of integers"""
    #Information about the implementation (not the abstraction)
    #Value of the set is represented by a list of ints, self.vals.
    #Each int in the set occurs in self.vals exactly once.
    def __init__(self):
        """Create an empty set of integers"""
        self.vals = []

    # THE FUNCTIONS BELOW ARE CALLED METHODS OR METHOD ATTRIBUTES OF THE CLASS
    # When an instance is made, we can invoke these methods using dot notation -- e.g., s.insert(2)
    def insert(self, e):
        """Assumes e is an integer and inserts e into self"""
        if e not in self.vals:
            self.vals.append(e)
    def member(self, e):
        """Assumes e is an integer
        Returns True if e is in self, and False otherwise"""
        return e in self.vals
    def remove(self, e):
        """Assumes e is an integer and removes e from self
        Raises ValueError if e is not in self"""
        try:
            self.vals.remove(e)
        except:
            raise ValueError(str(e) + ' not found')
    def getMembers(self):
        """Returns a list containing the elements of self.
        Nothing can be assumed about the order of the elements"""
        return self.vals[:]
    def __str__(self):
        """Returns a string representation of self"""
        # This is a special method. When an instance of this class is printed,
        # this function will be invoked and whatever is returned by this function
        # will be the output of the print
        self.vals.sort()
        result = ''
        for e in self.vals:
            result = result + str(e) + ','
        return '{' + result[:-1] + '}' #-1 omits trailing comma
```

In the above class definition, `self` is a formal parameter of type `IntSet()` that we can operate on as we define the methods/data attributes of the class.

When a new instance of type `IntSet` is created, the interpreter calls `__init__` and binds the newly created object to the formal parameter `self`. All data attributes inside this definition becomes part of the newly created object.

```python
s = IntSet()
s.vals #all instance will have the attribute .vals
```

Furthermore, when an instance is made, all functions inside the class definition can be invoked using dot notation.

```python
s.insert(3) # inserts the integer 3 to the instance `s`
s.vals
```

In the class definition, we may notice that we passed two parameters to the function `insert()`: the formal parameter `self` and an integer `e`, but passed only the integer `3` when we invoked the function using dot notation.

This is an artifact of the dot notation -- the object associated with the expression preceding the dot is also passed in the function called after the dot along with the other parameters we pass.

So, the expression `s.insert(3)` is equivalent to `insert(s,3)`.

```python
s.insert(2)
s.insert(1)
print(s)
```

## Using classes to keep track of students and faculty
Design a class that provides an abstraction of people, keeping track of their name and birthdays.

In particular, it should contain the data attributes `name`,`lastName` and `birthday`:

The class should contain the following methods:
- `getName`: returns the name of the person

- `getAge`: returns the age of the person
- `__lt__`: a way to order the class by last names, e.g., `Artoza` <  `Babol` < `Brutas`
- `__str__`: returns the output when the person is passed in the `print()` function

```python
import datetime
from dateutil import parser
import math

class Person(object):
    """A Person is a person identified by a name and a birthday"""
    
    def __init__(self, name):
        """Create person identified by specified name"""
        self.name = name
        try:
            lastBlank = name.rindex(' ')
            self.lastName = name[lastBlank+1:]
        except:
            self.lastName = name
        self.birthday = None

    def getName(self):
        """returns the name of the person"""
        return self.name

    def getLastName(self):
        """returns the last name of the person"""
        return self.lastName

    def setBirthday(self, birthdate):
        """Assumes birthdate is type str, containing a date
        sets the birthday of the person based on the specified birthdate"""
        self.birthday = birthdate

    def getAge(self):
        """returns the age of the person"""
        try:
            return math.floor(abs((parser.parse(self.birthday) - datetime.datetime.now()).days)/365)
        except:
            raise ValueError(f"No birthdate records for {self.name}")
            
    def __lt__(self, other):
        """
        Returns True if self precedes other in alphabetical order, and False otherwise.
        Comparison is based on last name, but if these are the same full names are compared.
        """
        try:
            if self.lastName == other.lastName:
                return self.name < other.name
            return self.lastName < other.lastName
        except:
            raise TypeError(
                f'\'<\' not supported between instances of \'{type(self).__name__}\' and \'{type(other).__name__}\''
            )
            
    def __str__(self):
        """Return self's name"""
        return self.name
```

The method `__lt__` overloads the `<` (less than) operator, allowing users to compare two objects of type `Person` based on their names. Aside from that, we can use other polymorphisms that use `__lt__`. One such method is `sort()`.

```python
nlist = ['Phil Dunphy', 'Jay Pritchet', 'Manny Delgado', 'Claire Dunphy']
plist = []
for name in nlist:
    plist.append(Person(name))
    print(name)

print('')
plist.sort()
for person in plist:
    print(person.name)
```
